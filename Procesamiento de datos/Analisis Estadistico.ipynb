```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python

def obtener_cantidad_videos_por_categoria(directorio_base):
    cantidad_videos_por_categoria = {}
    for categoria in os.listdir(directorio_base):
        ruta_categoria = os.path.join(directorio_base, categoria)
        if os.path.isdir(ruta_categoria):
            cantidad_videos = len(os.listdir(ruta_categoria))
            cantidad_videos_por_categoria[categoria] = cantidad_videos
    return cantidad_videos_por_categoria
```

```python

def generar_grafico_cantidad_videos_por_categoria(cantidad_videos_por_categoria):
    plt.figure(figsize=(10, 6))
    plt.bar(cantidad_videos_por_categoria.keys(), cantidad_videos_por_categoria.values(), color='skyblue')
    plt.xlabel('Categorías')
    plt.ylabel('Cantidad de Videos')
    plt.title('Cantidad de Videos por Categoría')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
def obtener_resoluciones_por_categoria(directorio_base):
    resoluciones_por_categoria = {}
    for categoria in os.listdir(directorio_base):
        ruta_categoria = os.path.join(directorio_base, categoria)
        if os.path.isdir(ruta_categoria):
            resoluciones = []  # Aquí deberías obtener las resoluciones de los videos en cada categoría
            resoluciones_por_categoria[categoria] = resoluciones
    return resoluciones_por_categoria
```

```python

def generar_histograma_resoluciones(resoluciones_por_categoria):
    plt.figure(figsize=(10, 6))
    for categoria, resoluciones in resoluciones_por_categoria.items():
        plt.hist(resoluciones, bins=20, alpha=0.5, label=categoria)
    plt.xlabel('Resolución')
    plt.ylabel('Cantidad de Videos')
    plt.title('Histograma de Resolución de Videos por Categoría')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

def obtener_duraciones_por_categoria(directorio_base):
    duraciones_por_categoria = {}
    for categoria in os.listdir(directorio_base):
        ruta_categoria = os.path.join(directorio_base, categoria)
        if os.path.isdir(ruta_categoria):
            duraciones = []  # Aquí deberías obtener las duraciones de los videos en cada categoría
            duraciones_por_categoria[categoria] = duraciones
    return duraciones_por_categoria

```

```python
def generar_diagrama_caja_duraciones(duraciones_por_categoria):
    plt.figure(figsize=(10, 6))
    plt.boxplot(duraciones_por_categoria.values(), labels=duraciones_por_categoria.keys())
    plt.xlabel('Categorías')
    plt.ylabel('Duración de Videos')
    plt.title('Diagrama de Caja de Duración de Videos por Categoría')
    plt.grid(True)
    plt.show()
```

```python
def generar_estadisticas_descriptivas_duraciones(duraciones_por_categoria):
    for categoria, duraciones in duraciones_por_categoria.items():
        print(f'Categoría: {categoria}')
        print(f'  Media: {np.mean(duraciones)}')
        print(f'  Mediana: {np.median(duraciones)}')
        print(f'  Mínimo: {np.min(duraciones)}')
        print(f'  Máximo: {np.max(duraciones)}')
        print('\n')
```

```python
# Obtener la ruta actual
ruta_actual = os.getcwd()

ruta_raiz= os.path.dirname(ruta_actual)


# Crear las carpetas 'dataset' y 'video' dentro de la ruta actual
ruta_dataset = os.path.join(ruta_raiz, 'Data_Set')
directorio_base = os.path.join(ruta_dataset, 'videos')

# Comprobar si las carpetas ya existen, si no, crearlas
if not os.path.exists(ruta_dataset):
    os.makedirs(ruta_dataset)

if not os.path.exists(directorio_base):
    os.makedirs(directorio_base)

# Obtener y mostrar cantidad de videos por categoría
cantidad_videos_por_categoria = obtener_cantidad_videos_por_categoria(directorio_base)
generar_grafico_cantidad_videos_por_categoria(cantidad_videos_por_categoria)
```
